```python
# Handle missing values
imputer = SimpleImputer(strategy='mean')

# One-hot encode categorical features
categorical_features = ['age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa', 'employment_industry', 'employment_occupation']
one_hot = OneHotEncoder(handle_unknown='ignore')

# Scale numerical features
numerical_features = ['xyz_concern', 'xyz_knowledge', 'behavioral_antiviral_meds', 'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal', 'chronic_med_condition', 'child_under_6_months', 'health_worker', 'health_insurance', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk', 'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults', 'household_children']
scaler = StandardScaler()

# Column transformer for features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[('imputer', imputer), ('scaler', scaler)]), numerical_features),
        ('cat', Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')), ('onehot', one_hot)]), categorical_features)
    ])

# Preprocess features
X_preprocessed = preprocessor.fit_transform(X)

# Split the data
X_train, X_val, y_train, y_val = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Initialize the model
model_xyz = RandomForestClassifier()
model_seasonal = RandomForestClassifier()

# Cross-validation
scores_xyz = cross_val_score(model_xyz, X_train, y_train['xyz_vaccine'], cv=5, scoring='roc_auc')
scores_seasonal = cross_val_score(model_seasonal, X_train, y_train['seasonal_vaccine'], cv=5, scoring='roc_auc')

print(f'Cross-validated ROC AUC scores for xyz_vaccine: {scores_xyz}')
print(f'Cross-validated ROC AUC scores for seasonal_vaccine: {scores_seasonal}')
print(f'Mean ROC AUC score for xyz_vaccine: {scores_xyz.mean()}')
print(f'Mean ROC AUC score for seasonal_vaccine: {scores_seasonal.mean()}')

# Train the model on full training data
model_xyz.fit(X_train, y_train['xyz_vaccine'])
model_seasonal.fit(X_train, y_train['seasonal_vaccine'])

# Predict probabilities on validation data
y_pred_xyz = model_xyz.predict_proba(X_val)[:, 1]
y_pred_seasonal = model_seasonal.predict_proba(X_val)[:, 1]

# Calculate ROC AUC scores
roc_auc_xyz = roc_auc_score(y_val['xyz_vaccine'], y_pred_xyz)
roc_auc_seasonal = roc_auc_score(y_val['seasonal_vaccine'], y_pred_seasonal)

# Mean ROC AUC score
mean_roc_auc = (roc_auc_xyz + roc_auc_seasonal) / 2

print(f'ROC AUC score for xyz_vaccine: {roc_auc_xyz}')
print(f'ROC AUC score for seasonal_vaccine: {roc_auc_seasonal}')
print(f'Mean ROC AUC score: {mean_roc_auc}')
```
